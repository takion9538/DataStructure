# **1. 스택이란?**

* 데이터를 제한적으로 접근할 수 있는 구조
* 한쪽 끝에서만 자료를 넣거나 뺄 수 있는 구조
* 가장 나중에 쌓은 데이터를 가장 먼저 빼낼 수 있는 구조
* LIFO(Last Input First Out)

### **1-1. 스택의 구조**

* 스택은 LIFO, FILO 방식의 데이터 관리
* 스택의 활용 : 컴퓨터 내부의 프로세스 구조의 함수 동작 방식
* 주소 기능
  * push() : 데이터를 스택에 쌓기
  * pop() : 마지막 데이터를 스택에서 꺼내기

### **1-2. 스택의 장단점**
* 장점
  1. 구조가 단순하여 구현이 쉬움
  2. 데이터 저장/읽기 속도가 빠름
* 단점
  1. 데이터 최대 개수를 미리 정의해야 한다(파이썬의 경우 재귀함수는  최대 1000번까지만 호출이 가능함)
  2. 저장공간의 낭비가 발생할 수 있음

> 스택은 단순하고 빠른 성능을 위해 사용하므로 보통 배열 구조를 활용하여 구현하는 것이 일반적이다

```python
# 파이썬 리스트 기능에서 제공하는 메소드로 스택 사용하기
data_stack = list()
print(data_stack)

# 데이터 추가
data_stack.append(1)
data_stack.append(10)
data_stack.append(5)
print(data_stack)

# 꺼내기
print(data_stack.pop())
print(data_stack)
```

### **문제**
리스트 변수로 스택을 다루는 pop, push 기능을 직접 구현한다. (pop, push 함순는 사용하지 않는다)

```python
stack_list = list()

# 데이터를 쌓는 함수
def push(data) : 
  stack_list.append(data)

# 마지막 데이터부터 추출하는 함수
def pop() : 
  datastack = stack_list[-1]
  del stack_list[-1]
  return datastack
  

push(1)
push(10)
push(5)
print(stack_list)

print(pop())
print(stack_list)
```
