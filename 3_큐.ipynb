# **1. 큐의 구조**

* 가장 먼저 넣은 데이터를 가장 먼저 꺼낼 수 있는 구조 (순차적 구조)
* 줄을 서는 행위와 유사하다.
* FIFO (first-in / forstout, 선입선출)


### **1-1. 큐의 용어**

- Enqueue : 큐에 데이터를 넣는 기능
- Dequeue : 큐에서 데이터를 꺼내는 기능
- [Visualgo] (http://visualgo.net/en/list)

### **1-2. 파이썬 queue 라이브러리를 활용하여 queue 자료구조 사용하기**

* Queue() : 가장 일반적인 큐 자료구조를 생성
* LifoQueue() : 나중에 입력된 데이터가 먼저 출력되는 구조의 큐 자료구조를 생성(스택구조와 비슷하다.)
* PriorityQueue() : 데이터마다 우선순위를 나누어 우선순위가 높은 순서대로 데이터를 출력한다.
* 프로그램을 작성할 때 프로그램에 따라서 적합한 자료구조를 사용한다.

```python
# Queue 를 사용하여 큐 생성

import queue

data_queue = queue.Queue()
data_queue.put('Hello') # EnQueue 기능
print(data_queue)
print(data_queue.qsize())
print(data_queue.get()) # Dequeue 기능
print(data_queue.qsize())

data_queue.put(10)
data_queue.put(True)

for i in range(data_queue.qsize()) :
  item = data_queue.get()
  print(item)
```

```python
# LifoQueue 를 사용하여 큐 생성

import queue

data_queue = queue.LifoQueue()
data_queue.put('Hello') # EnQueue 기능
print(data_queue)
print(data_queue.qsize())
print(data_queue.get()) # Dequeue 기능
print(data_queue.qsize())

data_queue.put(10)
data_queue.put(True)

for i in range(data_queue.qsize()) :
  item = data_queue.get()
  print(item)
```

```python
# PriorityQueue() 로 큐 생성

import queue

data_queue = queue.PriorityQueue()
data_queue.put((50, '고세구'))
data_queue.put((40, '아이네'))
data_queue.put((30, '징버거'))
data_queue.put((20, '주르르'))
data_queue.put((10, '릴파'))
print(data_queue.qsize())
print(data_queue.get()) # 키 값이 적을수록 우선순위가 높음
print(data_queue.get())
```

### **문제**
리스트 변수로 큐를 다루는 enqueue, dequeue 기능을 구현해보자

```python
import queue

queue_list = list()

def enqueue(data) :
  queue_list.append(data)

def dequeue() :
  data = queue_list[0]
  del queue_list[0]
  return data

enqueue(30)
enqueue(20)
enqueue(10)
print(queue_list)
print(dequeue())
print(dequeue())
print(queue_list)
```
