# **1. 더블 링크드 리스트(Doubly Linked List)**

* 양방향으로 연결되어 있어서 노드 탐색이 양쪽으로 모두 가능
* 한 데이터를 저장하기 위해 3 개의 데이터가 소모되기 때문에 리소스의 소모량이 크다

```python
class Node : 
  def __init__(self, data, prev=None, next=None) :
    self.prev = prev
    self.data = data
    self.next = next
```

```python
class NodeMgmt : 
  def __init__(self, data) :
    self.head = Node(data)
    self.tail = self.head

  def insert(self, data) : 
    if self.head == None :
      self.head = Node(data)
      return True
    if self.head == None : 
      self.head = Node(data)
      self.tail = self.head
    else : 
      node = self.head
      while node.next : 
        node = node.next
      new = Node(data)
      node.next = new
      new.prev = node
      self.tail = new
      
  def node_print(self) :
    node = self.head
    while node : 
      print(node.data)
      node = node.next

  def search_from_head(self, data) : 
    if self.head == None :
      self.head = Node(data)
      return True
    node = self.head
    if node.data == data : 
      print(node.data)
    else : 
      while node.next : 
        nodeNext = node.next
        if node.data == data :
          return data
        else : 
          node = node.next
      return False

  def search_from_tail(self, data) : 
    if self.head == None :
      self.head = Node(data)
      return True
    node = self.tail
    if node.data == data : 
      print(node.data)
    else : 
      while node.prev : 
        nodePrev = node.prev
        if node.data == data :
          return data
        else : 
          node = node.prev
      return False

  def delete(self, data) : 
    node = self.head
    if node.data == data : 
      self.head = node.next
      del node
    else : 
      while node.next : 
        nodeNext = node.next
        nodePrev = node.prev
        if nodeNext.data == data :
          node.next = nodeNext.next
          nodeNext.prev = nodePrev
          del node
          print('삭제되었습니다')
          return True
        else : 
          node = node.next
      return False

  def insert_before(self, data, before_data) : 
    if self.head == None : # 데이터가 삭제되었을 경우 self.head 조차 없을 수 있음
      self.head = Node(data)
      return True
    node = self.head
    if node.data == before_data : 
      new = Node(data)
      self.head = new
      node.prev = new
      new.next = node
    else : 
      while node.next : 
        nodePrev = node.prev
        if node.data == before_data :
          new = Node(data)
          new.prev = nodePrev
          new.next = node
          nodePrev.next = new
          node.prev = new
          return True
        else : 
          node = node.next
      return False
```

```python
doublelinkedlist1 = NodeMgmt(0)
for data in range(1, 11) :
  doublelinkedlist1.insert(data)
doublelinkedlist1.node_print()
```
