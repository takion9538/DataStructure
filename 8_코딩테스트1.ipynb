# **1. 좌표 정렬하기**
> https://www.acmicpc.net/problem/11650


```python
count = int(input('입력할 갯수를 입력 : '))

array = []

x = list(0 for i in range(count))
y = list(0 for i in range(count))
for _ in range(count) : # i 를 무시함
  x, y = map(int, input().split(' '))
  array.append((x, y))
  
array = sorted(array) # sort

for i in array : 
  print(i[0], i[1])
```

sort() 
* 리스트 메소드
* 원래 자료에 영향을 미쳐 변환한다.

sorted()
* 파이썬 내장 함수
* 새로운 정렬 결과를 리턴한다.
* 원래 자료에는 영향을 주지 않는다 > 새로 저장을 해주어야한다.

# **2. 소트 인사이드**
* https://www.acmicpc.net/problem/1427

```python
num = input()
len(num)

array = []

for i in range(len(num)) : 
  array.append(num[i])
array.sort(reverse = True)

for i in array : 
  print(i, end='')
```

```python
array = input()

for i in range(9, -1, -1):
  for j in array : 
    if int(j) == i : 
      print(j, end='')
```
